```python
d92diff = pd.read_csv('gcdiff.csv')
d92diff["eveWaveTime"]= pd.to_datetime(d92diff['eveWaveTime'],errors='coerce')
d92diff["mornWaveTime"]= pd.to_datetime(d92diff['mornWaveTime'],errors='coerce')
d92diff=d92diff.diff(periods= 1,axis = 0)
print(d92diff)
```

```python
d92diff['sec']=(d92diff['eveWaveTime'].dt.total_seconds())/60/60
d92diff['sec2']=(d92diff['mornWaveTime'].dt.total_seconds())/60/60

d92diff=d92diff.round(3)
print(d92diff)
```

```python
d92diff.to_csv('gcdiffhour.csv',encoding='utf_8_sig')
```
